# Final Practice

Read reddit_data.csv. Import pandas and datetime as pd and dt respectively.

```python
import pandas as pd
import datetime as dt

df = pd.read_csv('reddit_data.csv')
```

Have a look at the data. Each row represents a Reddit comment.

```python
df.head()
```

```python
df.shape
```

The utc columns shows the EPOCH timestamp format. We should perform data cleaning by replacing it.

```python
df['utc'] = pd.to_datetime(df['utc'],unit='s')
```

```python
df.head()
```

Let's create two new rows for storing the date and month.

```python
df['date'] = df['utc'].apply(lambda x: x.date())
```

```python
df['month'] = df['utc'].apply(lambda x: x.month)
```

```python
df.head()
```

Our client asks how many data are registered on December.

```python
df[df['month'] == 12].shape[0]
```

And how many data are after 1 January 2016 but only at odd months?

```python
df[(df['month'] % 2 == 1) & (df.utc > dt.datetime(2016, 1, 1))]
```

Now it's time to perform grouping. Display the top 10 users joining the most different subreddits.

```python
df.groupby('username')['subreddit'].nunique().sort_values(ascending = False).head(10)
df.groupby('username')[['subreddit']].nunique().sort_values(by = 'subreddit', ascending = False).head(10)
```

Final part, putting it altogether.
Find the top 10 most active redditors. The activity rate is measured from the ratio between number of comments and the number of unique subreddits joined.

```python
df_comments = df.groupby('username')[['subreddit']].count().rename(columns={'subreddit':'comments'})
df_subreddits = df.groupby('username')[['subreddit']].nunique().rename(columns={'subreddit':'unique_subs'})

df_comments.head()
```

```python
df_subreddits.head()
```

```python
df_combined = df_comments.merge(df_subreddits, left_index=True, right_index=True)
df_combined.head()
```

```python
df_combined['activity_rate'] = df_combined['comments']/df_combined['unique_subs']
df_combined.head()
```

Due to potential outliers, we only want to include those who has from 100 and 500 comments and from 12 to 30 unique subreddits.

```python
df_final = df_combined[(df_combined['comments'] >= 100) &
            (df_combined['comments'] <= 500) &
            (df_combined['unique_subs'] >= 12) &
            (df_combined['unique_subs'] <= 30)
            ]
```

```python
df_final = df_final.sort_values(by='activity_rate', ascending=False).head(10)
```

Finally, save the results into a new CSV file, feel free to think of a file name.

```python
df_final.to_csv('top10.csv', index = True)
```
