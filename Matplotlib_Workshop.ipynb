```python
def display_covid_data(countries):
  covid_full = pd.read_csv('covid_data.csv').loc[:,['Country/Region','Date','Confirmed','Deaths','Recovered','Active']]
  covid_full.Date = covid_full.Date.apply(pd.to_datetime)

  fig, ax = plt.subplots(2,2, figsize = (16,9))
  
  for c in countries:
    new_covid = covid_full[covid_full['Country/Region'] == c]
    ax[0,0].plot(new_covid.Date,new_covid.Confirmed,label=c)
    ax[0,0].set(xlabel='Date',title='Confirmed Cases')

    ax[0,1].plot(new_covid.Date,new_covid.Deaths,label=c)
    ax[0,1].set(xlabel='Date',title='Number of Deaths')

    ax[1,0].plot(new_covid.Date,new_covid.Recovered,label=c)
    ax[1,0].set(xlabel='Date',title='Number of Recovered')

    ax[1,1].plot(new_covid.Date,new_covid.Active,label=c)
    ax[1,1].set(xlabel='Date',title='Active Cases')
  
  for i in [0,1]:
    for j in [0,1]:
      ax[i,j].legend(loc='upper left')

display_covid_data(['Indonesia','Malaysia','Singapore'])
```

```python
# Part 2 : Simple Plot
x = tuple(range(100))
y = tuple(map(lambda x:x*x,x))

fig, ax = plt.subplots()

ax.plot(x,y)
ax.set_title(rf"Part 2 : Graph of $x^2$")
fig.show()
```

```python
# Part 3 : Multiple Plots
z = [x/50 for x in range(100)]
fig, ax = plt.subplots()
ax.plot(z,z, label = 'linear')
ax.plot(z,[x**2 for x in z], '.-', label = 'quadratic')
ax.plot(z,[x**3 for x in z], 'r.-', label = 'cubic')

ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title("Part 3")

ax.legend()
fig.show()
```

```python
# Part 4 : Dealing with raw data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
passengers = {1949:dict(zip(months,[112,118,132,129,121,135,148,148,136,119,104,118])),
              1950:dict(zip(months,[115,126,141,135,125,149,170,170,158,133,114,140])),
              1951:dict(zip(months,[145,150,178,163,172,178,199,199,184,162,146,166])),
              1952:dict(zip(months,[171,180,193,181,183,218,230,242,209,191,172,194])),
              1953:dict(zip(months,[196,196,236,235,229,243,264,272,237,211,180,201])),
              }

"""
dict(zip(['A','B','C'],[1,2,3])) = {'A':1, 'B':2, 'C':3}
"""

fig, ax = plt.subplots()

for year,row in passengers.items():
    ax.plot(list(row.keys()),list(row.values()),label=year)
ax.legend()
ax.set_title("Part 4 : Raw Data")
fig.show()
```

```python
# Part 5 : Boxplot
table = {}
for year in passengers:
    for month in passengers[year].keys():
        if month not in table:
            table[month] = []
        table[month].append(passengers[year][month])

"""
{"Jan":[112,113,114],"Feb":[100,101,122]}
"""

fig, ax = plt.subplots()
ax.boxplot(table.values(), labels = table.keys())
ax.set_title("Part 5 : Boxplot")
fig.show()
```

```python
# Part 6 : Stacked Bar
midterm = [['ID','Q1','Q2','Q3','Q4','Q5'],
           ['E1',5,1,5,5,2],
           ['E2',3,1,2,3,4],
           ['E3',0,1,2,0,4],
           ['E4',1,2,5,3,3],
           ['E5',0,0,0,0,5],
           ['E6',5,5,5,4,5],
           ['E7',5,5,5,4,5],
           ['E8',5,5,5,4,5],
           ['E9',5,5,5,4,5],
           ['E10',2,5,5,4,5],
           ]

head = midterm[0][1:] # ['Q1','Q2','Q3','Q4','Q5']
table = [[f'{i} marks']+[0]*len(head) for i in range(6)] # -> 6 x len(head) table

for row in midterm[1:]:
    for qn, marks in enumerate(row[1:]):
        table[marks][qn+1] += 1
    #print(list(enumerate(row[1:])))

fig, ax = plt.subplots()
ax.set_title("Part 6 : Stacked Bar")
#print(list(enumerate(map(lambda x:x[1:],table))))

btm = [0] * len(head)
for mark, row in enumerate(map(lambda x:x[1:],table)):
    ax.bar(head, row, bottom=btm, label=mark)
    btm = [x+y for x,y in zip(btm, row)]

ax.legend()
fig.show()
```
